```python
# TODO: Jupyter notebook demo explaining what can and can't be done

from pyspark import SparkContext

sc = SparkContext("local", "First App")


class Goo:
    def __init__(self, x):
        self.x = x

    def funcA(self):
        #print('x is {}'.format(self.x))
        #print(self.x)
        return self.x
goo1 = Goo(1)
goo2 = Goo(2)
goo3 = Goo(3)

def caller_strings(message):
    # goo instance
    if message== 'a':
        return goo1.funcA()
    elif message == 'b':
        return goo2.funcA()
    elif message == 'c':
        return goo3.funcA()

def instance_caller(goo_instance):
    return goo_instance.funcA()

#DOES_NOT_WORK = sc.parallelize([goo1, goo2, goo3]).map(instance_caller).take(3)
WORKS = sc.parallelize(['a', 'b', 'c']).map(caller_strings)
print(WORKS)
print(WORKS.take(5))

#print(instances.map(funcA_caller).take(3))
#print(nums.map(funcA_caller).take(3))
#print(sc.parallelize([1,2,3]).map(times_2_wrapper).take(2))

# TODO email lalana
```
